### Estudo dirigido 5 - Árvores de decisão

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, accuracy_score
```

Carregando o dataset

```python
from os import path, getcwd

cur_dir   = getcwd()
data_dir  = path.join(cur_dir, 'data')
csv_file  = path.join(data_dir, 'decision_tree.csv')
df        = pd.read_csv(csv_file)

df
```

Separando o dataset em `features` e `target`

```python
feature_cols = ['company', 'job', 'degree']
X = df[feature_cols]
Y = df['salary']
```

Codificar colunas utilizando labels

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

for f in feature_cols:
    
    X[f] = label_encoder.fit_transform(X[f])
    
X
```

Train test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

Criando o classificador


```python
entropy_classifier = DecisionTreeClassifier(criterion="entropy", max_depth=3)

entropy_classifier = entropy_classifier.fit(X_train, Y_train)

Y_pred = entropy_classifier.predict(X_test)

print(f"Accuracy score: {accuracy_score(Y_test, Y_pred)}")
```
